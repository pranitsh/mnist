The goal here is to create an efficient algorithmic way to determine what number an image represents.

````python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt

mnist_data = fetch_openml('mnist_784', version=1, return_X_y=True)
images, targets = mnist_data
images = images.to_numpy()
targets = targets.astype(int)


# Visualizing an iamge
image_idx = 0
an_image = images[image_idx][:].reshape(28, 28)
print("Displaying number", targets[image_idx])
plt.imshow(an_image) # Should display a 5 that looks like an S.
````

Let's display the images we will try to learn from.

````python
# Let's get the first index of each number
number_idx = {}
for idx, number in enumerate(targets):
  if number not in number_idx:
    number_idx[number] = idx

for number in range(10):
  print(number, ": ", number_idx[number])

# Modified from 3.5 AI code
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
for number in range(10):
  ax = axes[number // 5, number % 5]
  ax.imshow(images[number_idx[number]][:].reshape(28, 28), cmap='gray')
  ax.set_title(number)
  ax.axis('off')
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

def mean_squared_error(y_true, y_pred):
  return np.mean((y_true - y_pred) ** 2)

def min_probe(index):
  global number_idx
  global images
  probe_list = []
  for number in range(10):
    probe_list.append(mean_squared_error(images[number_idx[number]], images[index]))
  min = np.Infinity
  min_idx = None
  for idx, score in enumerate(probe_list):
    if score < min:
      min = score
      min_idx = idx
  return min_idx, probe_list

correct = 0
incorrect = []
for number in range(70000):
  prediction, approximations = min_probe(number)
  actual = targets[number]
  if prediction == actual:
    correct += 1
  else:
    incorrect.append((number, prediction, actual, approximations))
  
print("Error rate of", 100*(1-correct/70000), "%")

import random

random.seed(0)
shuffled_incorrects = incorrect[:]
random.shuffle(shuffled_incorrects)

amount_to_show = 20
fig, axes = plt.subplots(4, 5, figsize=(12, 6))

for idx, (image_idx, prediction, actual, approximations) in enumerate(shuffled_incorrects[:amount_to_show]):
  ax = axes[idx // 5, idx % 5]
  ax.imshow(images[image_idx][:].reshape(28, 28), cmap='gray')
  ax.set_title(f"Guessed {prediction}, Actual {actual}")
  ax.set_xlabel(f"{image_idx}")
  print(image_idx, ":", (approximations / np.sum(approximations)).astype('float16'))
plt.tight_layout()
plt.show()
````

An error rate of 48.167142857142853%. A good error rate is around 0.25%. Still, it was on only trained on 10 images, and it ran in approximately 10 seconds on CPU.

The next goal is to take an already working implementation that runs fast, and convert it to use a data structure-centered approach to improve efficiency.

````python
from sklearn.metrics import classification_report 
import time
sizes = 10000


start_time = time.time()
from sklearn.ensemble import RandomForestClassifier
randf_model = RandomForestClassifier(n_estimators=100, random_state=42)
randf_model.fit(images[:sizes], targets[:sizes])
randf_pred = randf_model.predict(images[sizes:2*sizes])
end_time = time.time()
print("Random Forest Regression", ":", end_time - start_time, "seconds")
randf_report = classification_report(randf_pred, targets[sizes:2*sizes], output_dict=True, digits=6)
print("Error Rate: ", 100*(1 - randf_report['accuracy']), "%")


start_time = time.time()
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(C=50.0, tol=0.01, 
                           penalty="l1", 
                           solver="saga") 
logistic_model.fit(images[:sizes], targets[:sizes]) 
logistic_pred = logistic_model.predict(images[sizes:2*sizes])
end_time = time.time()
print("Logistic Regression", ":", end_time - start_time, "seconds")
logistic_report = classification_report(logistic_pred, targets[sizes:2*sizes], output_dict=True, digits=6)
print("Error Rate: ", 100*(1 - logistic_report['accuracy']), "%")


start_time = time.time()
from sklearn.svm import SVC
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_model.fit(images[:sizes], targets[:sizes])
svm_pred = svm_model.predict(images[sizes:2*sizes])
end_time = time.time()
print("Support Vector Machine Classification", ":", end_time - start_time, "seconds")
svm_report = classification_report(svm_pred, targets[sizes:2*sizes], output_dict=True, digits=6)
print("Error Rate: ", 100*(1 - svm_report['accuracy']), "%")


start_time = time.time()
from sklearn.ensemble import GradientBoostingClassifier
gbm_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gbm_model.fit(images[:sizes], targets[:sizes])
gbm_pred = gbm_model.predict(images[sizes:2*sizes])
end_time = time.time()
print("Gradient Boosting Machines Classification", ":", end_time - start_time, "seconds")
gbm_report = classification_report(gbm_pred, targets[sizes:2*sizes], output_dict=True, digits=6)
print("Error Rate: ", 100*(1 - gbm_report['accuracy']), "%")


start_time = time.time()
from sklearn.neighbors import KNeighborsClassifier
kn_model = KNeighborsClassifier(n_neighbors=5)
kn_model.fit(images[:sizes], targets[:sizes])
kn_pred = kn_model.predict(images[sizes:2*sizes])
end_time = time.time()
print("K Neighbors Classifiers", ":", end_time - start_time, "seconds")
kn_report = classification_report(kn_pred, targets[sizes:2*sizes], output_dict=True, digits=6)
print("Error Rate: ", 100*(1 - kn_report['accuracy']), "%")
````

The one that performed the best in my mind is K neighbors classifier. It's speed was the fastest and the accuracy is still fairly high.

````python
import numpy as np
import time

class KNN_Classifier:
  def __init__(self, x_train, x_true, k):
    self.x_train = x_train
    self.x_true = x_true
    self.k = k

  def predict(self, vector):
    distances = np.linalg.norm(self.x_train - vector[None, :], axis=-1)
    kth_smallest = np.argpartition(distances, self.k - 1)[:self.k]
    count = [0]*10
    for idx in kth_smallest:
      count[self.x_true[idx]] += 1
    return np.argmax(count)


sizes = 10000
test_size = 1000
correct = 0

start_time = time.time()
kn_model = KNN_Classifier(images[:sizes], targets[:sizes], 5)
for image_idx in range(10000, 10000 + test_size):
  kn_pred = kn_model.predict(images[image_idx])
  # print(image_idx, end="")
  if kn_pred == targets[image_idx]:
    correct += 1
end_time = time.time()
print("K Neighbors Classifiers", ":", end_time - start_time, "seconds")
print("Error Rate: ", 100 * (test_size - correct) / test_size, "%")
````
